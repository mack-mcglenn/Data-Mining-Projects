#### Data Visualization

There are a few relationships I'm interested in exploring. First, I am curious if there are similarities or differences in how much study time different majors commit to. To see this relationship, I need to know how many categorical majors there are. I used value_counts to get that information. Next, I used a scaterplot to observe the data. I assigned each major a color to clearly differentiate them. I used [this link](https://matplotlib.org/stable/gallery/color/named_colors.html) to find different colors to assign.

```python
filtered_studentdf.major.value_counts()
```

```python
# Using matplotlib for initial visualization

fig, ax = plt.subplots()
colors = {'Arts':'brown',
          'Psychology':'pink',
          'Computer Science':'black', 
          'Business':'olive',
          'Engineering':'cyan',
          'Biology':'red'}

#mapping majors by color
color=filtered_studentdf['major'].map(colors)

# scatterplot exploring relationship between different majors and time spent studying
ax.scatter(filtered_studentdf['exam_score'], filtered_studentdf['study_hours_per_day'], c=color)

# set a title and labels
ax.set_title('Study Length and Exam Scores by Major')
ax.set_xlabel('exam_score')
ax.set_ylabel('study_hours_per_day')
plt.show()
```

This initial scatterplot doesn't show me any coherent information, so I think it'll be more useful to look at each major independently

```python
majors = filtered_studentdf['major'].unique()

for major in majors:
    data = filtered_studentdf[filtered_studentdf['major'] == major]
    plt.figure(figsize=(8,6))
    plt.scatter(data['exam_score'], data['study_hours_per_day'])
    plt.title(f"{major}")
    plt.xlabel('Exam Score')
    plt.ylabel('Study Hours per Day')
    plt.show()
    #input("Press Enter to continue to next major...")
```

This also isn't quite what I'm looking for. There's too much data and it's overwhelming, so I'm going to try a violin plot instead

```python
sb.violinplot(x='major', y='study_hours_per_day', data=filtered_studentdf)
plt.title('Individual Study Hours by Major')
plt.xlabel('Major')
plt.ylabel('Study Hours per Day')
plt.show()
```

```python
Now I can clearly see the information. According to the findings, computer science majors spend the highest average of study time
```
